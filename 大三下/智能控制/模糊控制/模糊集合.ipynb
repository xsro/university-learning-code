# 计算模糊集合的隶属度

智能控制老师在群里调查了一下大家认为的青年人的年龄范围，
我觉得可以用这个作为数据计算一下大家认为的青年人的隶属度

## 计算

首先，使用正则表达式提取聊天记录中的信息

```python
import re
#读取文件信息
QQchatData = open("data.txt", "r",encoding = "utf-8").read()
#使用正则表达式进行匹配,提取数据
pattern=re.compile(r"(B.*?)\d+:\d+:\d+\n(\d+)[-|—|－|~|_|-]{1,2}(\d+)")
m=pattern.findall(QQchatData)
print("收集到数据{}条".format(len(m)))
```

接着，计算隶属度

```python
import numpy as np
ls=[]
#
for item in m:
    min=int(item[1])
    max=int(item[2])
    x=min
    while x<=max:
        ls.append(x)
        x=x+1
#统计认为某岁属于青年的人数
x=[];y=[]
for i in set(ls):
    x.append(i)
    y.append(ls.count(i))
```

绘图，没有做归一化

```python
import matplotlib.pyplot as plt

plt.plot(x,y,'o-')
plt.grid()

plt.show()
```
